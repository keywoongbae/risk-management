```r
# library import
library("fitdistrplus")
library("actuar")
```

<hr>

```r
# Fitness of Discrete Distribution
data("toxocara")
str(toxocara)
toxocara$number
```

```r
# Fitness to Poisson and Negative-Binomial Distribution
toxo.P <- fitdist(toxocara$number, "pois")
toxo.nb <- fitdist(toxocara$number, "nbinom")
```

```r
# draw pdf, cdf, qqplot, ppplot
par(mfrow = c(2,2))
plot.legend <- c("Poisson", "Negative-Binomial")

denscomp(list(toxo.P, toxo.nb), legend = plot.legend)
qqcomp(list(toxo.P, toxo.nb), legend = plot.legend)
cdfcomp(list(toxo.P, toxo.nb), legend = plot.legend)
ppcomp(list(toxo.P, toxo.nb), legend = plot.legend)
```

```r
gofstat(list(toxo.P, toxo.nb), fitnames = c("pois","nbinom"))
```

```r
data("groundbeef")
str(groundbeef)
```

```r
# plot a density
plotdist(groundbeef$serving, histo = TRUE, demp = TRUE)
```

```r
# Fitness of continous distribution
fw <- fitdist(groundbeef$serving, "weibull")
fg <- fitdist(groundbeef$serving, "gamma")
fln <- fitdist(groundbeef$serving, "lnorm")
```

```r
par(mfrow = c(2,2))
plot.legend <- c("Weibull", "Lognormal", "Gamma")

denscomp(list(fw, fln, fg), legendtext = plot.legend)
qqcomp(list(fw, fln, fg), legendtext = plot.legend)
cdfcomp(list(fw, fln, fg), legendtext = plot.legend)
ppcomp(list(fw, fln, fg), legendtext = plot.legend)
```

```r
gofstat(list(fw, fln, fg), fitnames = c("Weibull", "Lognormal", "Gamma"))
```
